```python
import pandas as pd
base = pd.read_csv('houses_price.csv')
base.head()
```

```python
import numpy as np

import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
# columns used in training
used_columns = [ 'price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']

base = pd.read_csv('houses_price.csv', usecols=used_columns) 
```

```python
# import normalization
from sklearn.preprocessing import MinMaxScaler

scaler_x = MinMaxScaler()

# Chose attributes
used_attrs = ['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']

# scaler attributes
base[used_attrs] = scaler_x.fit_transform(
    base[used_attrs]
)

base.head()
```

```python
scaler_y = MinMaxScaler()

# scallered only the classe prevision (price), previsors and classes are scallered individual
attr_y = ['price']
base[attr_y] = scaler_y.fit_transform(base[attr_y])
base.head()
```

```python
axis_x = base.drop('price', axis=1)
axis_y = base.price
# dataframe
axis_x.head()
# transform y (prices), in numpy series
# axis_y.head()
```

```python
# took all the features, less the attribute price 
previsors_columns = used_columns[1::]

# create column for tensorflow
columns = [tf.feature_column.numeric_column(key=val) for val in previsors_columns]
print(columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    axis_x,
    axis_y, 
    test_size=0.3
)
```

```python
function_train = tf.estimator.inputs.pandas_input_fn(
    x = x_train, 
    y =  y_train,
    batch_size=32,
    num_epochs=None,
    shuffle=True
)

function_test = tf.estimator.inputs.pandas_input_fn(
    x = x_test, 
    y =  y_test,
    batch_size=32,
    num_epochs=10000,
    shuffle=False
)
```

```python
# instanciate the linear regressor algorithm
regressor = tf.estimator.LinearRegressor(
    feature_columns=columns
)
```

```python
# training the model
regressor.train(
    input_fn=function_train,
    steps=10000
)
```

```python
# evaluate the model
metrics_traning = regressor.evaluate(
    input_fn=function_train,
    steps=10000
)

metrics_traning
```

```python
metrics_test = regressor.evaluate(
    input_fn=function_test,
    steps=10000
)
metrics_test
```

```python
# =====> predictions
import numpy as np

function_predict = tf.estimator.inputs.pandas_input_fn(x = x_test, shuffle=False)
previsions = regressor.predict(
    input_fn=function_predict
)
# cast to list
values_predict = []
for p in previsions:
    values_predict.append(p['predictions'])

# transform in matrix
values_predict = np.asarray(values_predict).reshape(-1, 1)    

# receive a matrix for make the inversion (must be an array reshap)
values_real_predicted = scaler_y.inverse_transform(values_predict)
values_real_predicted
```

```python
y_test = np.asarray(y_test).reshape(-1, 1)
y_test
value_real_ytest = scaler_y.inverse_transform(y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(value_real_ytest, values_real_predicted)
mae
# error dicrease compare with a unique label
```
