```
# regression
```

```
dataframe_training1 = dataframe[["hum", "cnt", "casual"]]
```

```
dataframe_training1["hum"].fillna(dataframe_training1["hum"].mean(), inplace=True)
dataframe_training1["cnt"].fillna(dataframe_training1["cnt"].mean(), inplace=True)
dataframe_training1["casual"].fillna(dataframe_training1["casual"].mean(), inplace=True)

dataframe_training1.info()
```

```
dataframe_training1.isnull().sum()
```

```
x1 = dataframe_training1[["hum", "casual"]]
y1 = dataframe_training1[["cnt"]]
```

```
from sklearn.linear_model import LinearRegression

linearRegression = LinearRegression()
linearRegression.fit(x1, y1)
```

```
linearRegression.coef_
```

```
linearRegression.intercept_
```

```
 y1_pred = linearRegression.predict(x1)
 from sklearn.metrics import r2_score
 r2_score(y1_pred, y1)
```

```
 # decision tree
```

```
from sklearn import tree

decTree = tree.DecisionTreeRegressor()
decTree.fit(x1, y1)
```

```
y2_pred = decTree.predict(x1)
r2_score(y2_pred, y1)
```
